```python
def get_statistics(input_list):
    # Write your code here.
    return {
        "mean": cal_mean(input_list),
        "median": cal_median(input_list),
        "mode": cal_mode(input_list),
        "sample_variance": cal_variance(input_list),
        "sample_standard_deviation": cal_std(input_list),
        "mean_confidence_interval": cal_confi_interv(input_list),
    }
def cal_mean(input_values):
    if len(input_values) == 0:
        return 0 
    return sum(input_values) / len(input_values)   

def cal_mode(input_values):
    dictionary={} # Create an empty dictionary to store
    for value in input_values:
        if value not in dictionary:
            dictionary[value]=1
        else:
            dictionary[value] +=1

    max_num= float("-inf")
    max_num_key= None

    for key, value in dictionary.items():
        if value > max_num:
            max_num = value
            max_num_key=key
    return max_num_key 

def cal_median(input_values):
    sorted_values = sorted(input_values)
    if len(sorted_values) % 2 == 1:
        return sorted_values[len(input_values) //2]
    else:
        return (sorted_values[len(input_values) //2] + sorted_values[(len(input_values) //2)-1]) / 2 
#def cal_mode(input_list):
    #mode = max(set(input_list), key=input_list.count)
    #return mode
def cal_variance(input_values):
    return cal_std(input_values) ** 2


def cal_std(input_values):
    std = 0
    mean = cal_mean(input_values)
    
    for value in input_values:
        std += (value - mean) ** 2
    return (std / (len(input_values) - 1)) ** 0.5
        



def cal_confi_interv(input_values):
    std = cal_std(input_values)
    mean = cal_mean(input_values)
    x = 1.96
    temp = x * (std/(len(input_values)**0.5))
    return [mean - temp, mean + temp]
```

## Solution [2]

```python
def get_statistics(input_values):
    # Write your code here.
    input_values= sorted(input_values) #This refers to the list of values that we want to find
    num = len(input_values)
    mean = sum(input_values) / num
    middle_num = (num - 1) // 2 # // floor division
    #an integer that represents the index of the middle element in the sorted input_values list.

    median = (input_values[middle_num] + input_values[~middle_num]) / 2
    
    mode = max(set(input_values), key= input_values.count)
    
    Variance_sample = sum([(numx - mean)**2/(num-1) for numx in input_values])
    Standard_devation = Variance_sample **0.5

    std_error = Standard_devation / (num)**0.5
    Z_score = 1.96 * std_error
    mean_confidence_interval = [mean - Z_score, mean + Z_score]

    

    
    return {
        "mean": mean,
        "median": median,
        "mode": mode,
        "sample_variance": Variance_sample,
        "sample_standard_deviation": Standard_devation,
        "mean_confidence_interval": mean_confidence_interval,
    }
```
